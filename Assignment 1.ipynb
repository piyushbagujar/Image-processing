```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("/content/nature.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray')
plt.title("Original Image")
plt.axis("off")
```

```python
bright_image_improve=cv2.add(img,80)
plt.imshow(bright_image_improve,cmap='gray')
plt.title("Brightness Improvement")
plt.axis("off")
```

```python
bright_image_reduction=cv2.subtract(img,100)
plt.imshow(bright_image_reduction,cmap='gray')
plt.title("Brightness Reduction")
plt.axis("off")
```

```python
T = 127   # threshold value
_, thresh_img = cv2.threshold(img, T, 255, cv2.THRESH_BINARY)

plt.imshow(thresh_img,cmap='gray')
plt.title("Thresholding (cv2.threshold)")
plt.axis("off")
plt.show()
```

```python
neg_img = 255 - img
plt.imshow(neg_img, cmap='gray')
plt.title("Negative Image Using Formula")
plt.axis("off")
plt.show()
```

```python
neg_img_cv = cv2.bitwise_not(img)

plt.imshow(neg_img_cv, cmap='gray')
plt.title("Negative Using bitwise_not")
plt.axis("off")
plt.show()
```

```python
img_float = img.astype(float)

# Constant for scaling
c = 255 / np.log(1 + np.max(img_float))

# Apply log transformation
log_img_float = c * np.log(1 + img_float)

log_img_uint8 = np.array(log_img_float, dtype=np.uint8)

plt.imshow(log_img_uint8, cmap="gray")
plt.title("Log Transform (8-bit)")
plt.axis("off")
plt.show()
```

```python
gamma = 0.5

# Convert to float [0,1]
img_float = img.astype(np.float32) / 255.0

# Apply gamma correction
power_law = cv2.pow(img_float, gamma)

# Convert back to [0,255]
power_law = np.uint8(power_law * 255)

plt.imshow(power_law, cmap='gray')
plt.title("Power Law (cv2.pow)")
plt.axis("off")
```
